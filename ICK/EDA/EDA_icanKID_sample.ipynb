```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.figure_factory as ff

import os
from datetime import datetime

chart_color = px.colors.qualitative.Plotly
```

```python
if os.name == 'nt':
    dataframe_path = ''
else:
    dataframe_path = '/Users/admin/_Work/Data/icanKID/Data_sample_140423.csv'
df = pd.read_csv(dataframe_path)
```

```python
display(df.info())
display(df.head())
```

```python
df['UserJoinedDate'] = pd.to_datetime(df['UserJoinedDate']).dt.strftime(r'%Y-%m-%d')
df['Event_dayofweek'] = pd.to_datetime(df['EventDate']).dt.dayofweek
day_of_week = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
dict_dayofweek = {i:v for i,v in enumerate(day_of_week)}
df['Event_dayofweek'] = df['Event_dayofweek'].apply(lambda x: dict_dayofweek[x])

# df['EventDate'] = pd.to_datetime(df['EventDate']).dt.strftime(r'%Y-%m')
pay_user = df[df['gapday'].notna()]
free_user = df[df['gapday'].isna()]
```

```python
df[df['UserID']==pay_user['UserID'].iloc[0]].head(15)
```

```python
temp = df.copy()
temp = pd.pivot_table(temp, index='EventDate', values='nth_day_from_registration', aggfunc='count')
fig = px.histogram(temp,
    x=temp.index,
    y=temp['nth_day_from_registration'],
    title= 'Days from registration' 
                )
fig.show()
```

```python
pay_user['payday'] = pd.to_datetime(pay_user['UserJoinedDate']) + pay_user['gapday'].apply(np.ceil).apply(lambda x: pd.Timedelta(x, unit='D'))
pay_user['pay_dayofweek'] = pay_user['payday'].dt.dayofweek
pay_user['joined_dayofweek'] = pd.to_datetime(pay_user['UserJoinedDate']).dt.dayofweek
day_of_week = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
dict_dayofweek = {i:v for i,v in enumerate(day_of_week)}
pay_user['pay_dayofweek'] = pay_user['pay_dayofweek'].apply(lambda x: dict_dayofweek[x])
pay_user['joined_dayofweek'] = pay_user['joined_dayofweek'].apply(lambda x: dict_dayofweek[x])
pay_user.head(5)
```

```python
temp = pay_user['pay_dayofweek'].value_counts(sort=True)
temp2 = pay_user['joined_dayofweek'].value_counts(sort=True)
temp3 = pay_user['Event_dayofweek'].value_counts(sort=True)
temp_df = pd.DataFrame([temp, temp2, temp3]).transpose()

fig = px.bar(temp_df, x=temp_df.index, y=['pay_dayofweek', 'joined_dayofweek', 'Event_dayofweek'], title='Days of week event occur',barmode='group')
fig.update_xaxes(categoryorder='array', categoryarray=day_of_week, title='Day of week')
fig.update_yaxes(title='count')
fig.show()
```

```python
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=['Pay user', 'Free user'], shared_yaxes=True
    )

df_show = [pay_user, free_user]
for i, df_ in enumerate(df_show):
    for j, day in enumerate(day_of_week):
        fig.add_trace(
            go.Box(
            y=df_[df_['Event_dayofweek'] ==day]['UsageTime'], 
            name=day,
            marker_color= chart_color[j if j < len(chart_color) else (j - round(j/len(chart_color)) * len(chart_color))],
            legendgroup=day,
            showlegend=False if i >0 else True),
            row=1, col=i + 1
        )
fig.show()
```

```python
UsageTime_ActionType = pd.pivot_table(pay_user, index='nth_day_from_registration', columns='ActionType', values='UsageTime', aggfunc=['count',np.mean, np.sum, np.std])
OpenTime_ActionType = pd.pivot_table(pay_user, index='nth_day_from_registration', columns='ActionType', values='open_time', aggfunc=['count',np.mean, np.sum, np.std])
Records_ActionType = pd.pivot_table(pay_user, index='nth_day_from_registration', columns='ActionType', values='Records', aggfunc=['count',np.mean, np.sum, np.std])
```

```python

list_of_action = df['ActionType'].unique()
x_values = sorted(df['nth_day_from_registration'].unique())
attributes = [x for x in np.unique([x[0] for x in UsageTime_ActionType.columns]) if x!= 'std']
print(attributes)
fig = make_subplots(
    rows=len(attributes), cols=1,
    subplot_titles=[f'Pay user Usage Time by: {i}' for i in attributes], 
    # shared_yaxes=True,
    vertical_spacing = 0.1
    )

flag = 0
for i, attribute in enumerate(attributes):
    for j, action in enumerate(list_of_action):
        fig.add_trace(
            go.Scatter(
            x=x_values[:201],
            y=Records_ActionType[attribute][action][:201], 
            name=action,
            mode='lines',
            marker_color= chart_color[j if j < len(chart_color) else (j - round(j/len(chart_color)) * len(chart_color))],
            legendgroup=action, 
            # error_y = dict(type='data', color = chart_color[i],array=UsageTime_ActionType['std'][action], visible=True),
            showlegend=False if flag > 0 else True),
            row=i +1, col=1
        )
    flag += 1
fig.update_xaxes(title='N-th day from registration')
fig.update_layout(height=1500)
fig.show()
```

```python
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=['Pay user', 'Free user'], shared_yaxes=True
    )
list_of_action = df['ActionType'].unique()
flag = 0
df_show = [pay_user, free_user]
for i, df_ in enumerate(df_show):
    for j, action in enumerate(list_of_action):
        fig.add_trace(
            go.Bar(
            x=day_of_week,
            y=[df_[(df_['Event_dayofweek'] ==day) & (df_['ActionType']==action)]['EventDate'].count()
                for day in day_of_week], 
            name=action,
            marker_color= chart_color[j if j < len(chart_color) else (j - round(j/len(chart_color)) * len(chart_color))],
            legendgroup=action,
            showlegend=False if flag >0 else True),
            row=i+1, col=1
        )
    flag += 1
fig.update_layout(title='Event frequency of 3 ActionType')
fig.show()
```
